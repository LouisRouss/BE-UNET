```python
import torch
import torch.nn as nn
import numpy as np
import random
import matplotlib.pyplot as plt
import os,sys
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, datasets, models
import torchvision.utils
from collections import defaultdict
import torch.nn.functional as F
import torch.optim as optim
from torch.optim import lr_scheduler
import time
import copy
from functools import reduce
from torch.autograd import Variable
import itertools
import tifffile
```

```python
import albumentations as A
import albumentations.augmentations.transforms as AF
from albumentations.pytorch import ToTensorV2
```

```python
class ISBI2012Dataset(Dataset):
    
    def __init__(self, path_img, path_target, transforms=None):
        self.train = np.expand_dims(tifffile.TiffFile(path_img).asarray(), axis=-1)
        self.targets = np.expand_dims(tifffile.TiffFile(path_target).asarray(), axis=-1)
        self.transforms = transforms
        
    def __len__(self):
        return len(self.train)
    
    def __getitem__(self, idx):
        img = self.train[idx]
        target = self.targets[idx]
        
        if self.transforms:
            transformed = self.transforms(image=img,mask=target)
            img = transformed["image"]
            target = transformed["mask"]
            
        return img, target
```

```python
transform = A.Compose([AF.ElasticTransform(),
                       AF.HorizontalFlip(),
                       AF.VerticalFlip(),
                       AF.ColorJitter(),
                       AF.GaussNoise(),
                       AF.RandomBrightnessContrast(),
                       A.pytorch.ToTensorV2()
                      ])
isbi = ISBI2012Dataset('./dataset/train-volume.tif', './dataset/train-labels.tif',
                       transforms=transform)
```

```python
class double_conv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(double_conv, self).__init__()
        self.batch_norm = nn.BatchNorm2d(out_channels)
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1)
        
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.batch_norm(x)
        x = F.relu(self.conv2(x))
        x = self.batch_norm(x)
        return x
```

```python
class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        self.pool1 = nn.MaxPool2d(kernel_size=2)
        self.pool2 = nn.MaxPool2d(kernel_size=2)
        self.pool3 = nn.MaxPool2d(kernel_size=2)
        self.pool4 = nn.MaxPool2d(kernel_size=2)
        self.pool5 = nn.MaxPool2d(kernel_size=2)
        
        self.up1 = nn.Upsample(scale_factor=2, mode='nearest')
        self.up2 = nn.Upsample(scale_factor=2, mode='nearest')
        self.up3 = nn.Upsample(scale_factor=2, mode='nearest')
        self.up4 = nn.Upsample(scale_factor=2, mode='nearest')
        self.up5 = nn.Upsample(scale_factor=2, mode='nearest')
        
        self.conv1 = double_conv(1, 32)
        self.conv2 = double_conv(32, 64)
        self.conv3 = double_conv(64, 128)
        self.conv4 = double_conv(128, 256)
        self.conv5 = double_conv(256, 512)
        self.conv51 = double_conv(512,1024)
        
        self.conv6 = double_conv(1536,512)
        self.conv7 = double_conv(768, 256)
        self.conv8 = double_conv(384, 128)
        self.conv9 = double_conv(192, 64)
        self.conv10 = double_conv(96, 32)
        
        self.conv11 = nn.Conv2d(32, 1, 1)
        self.output = nn.Sigmoid()
        
    def forward(self, x):
        c1 = self.conv1(x)
        x = self.pool1(c1)
        c2 = self.conv2(x)
        x = self.pool2(c2)
        c3 = self.conv3(x)
        x = self.pool3(c3)
        c4 = self.conv4(x)
        x = self.pool4(c4)
        c5 = self.conv5(x)
        x = self.pool5(c5)
        x = self.conv51(x)
        x = self.up1(x)
        x = torch.cat([x, c5], 1)
        x = self.conv6(x)
        x = self.up2(x)
        x = torch.cat([x, c4], 1)
        x = self.conv7(x)
        x = self.up3(x)
        x = torch.cat([x, c3], 1)
        x = self.conv8(x)
        x = self.up4(x)
        x = torch.cat([x, c2], 1)
        x = self.conv9(x)
        x = self.up5(x)
        x = torch.cat([x,c1],1)
        x = self.conv10(x)
        x = self.conv11(x)
        #x = self.output(x)
        return x
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = UNet()
model = model.to(device)
```

```python
def train_model(model,dataset,criterion, optimizer, n_epoch=1, batch_size=1, shuffle=False):
    
    loss_history = []
    
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)
    
    for epoch in range(n_epoch):
        current_loss = 0
        
        print('Epoch : {}/{}'.format(epoch+1,n_epoch))
        print('*'*10)
        
        for batch, (data,target) in enumerate(loader):
            #data, target = Variable(data.cuda()) , Variable(target.cuda())
            
            data = data.to(device=device, dtype=torch.float)
            target = target.to(device=device, dtype = torch.float)
            data, target = data.cuda(), target.cuda()
            optimizer.zero_grad()
            output = model(data)
            output = output.permute(0,2,3,1)
            target = target/255
            loss = criterion(output,target)
            current_loss += loss.item()
            
            if (batch%10) == 9:
                print('\tBatch : {}/{}\tLoss : {:.4f}'.format(batch+1, len(loader), loss.item()))
            
            loss.backward()
            optimizer.step()
            
        loss_history.append(current_loss/len(loader))
    return model,loss_history

def predict(model,data):
        if isinstance(data, np.ndarray):
            data = torch.from_numpy(data)
        data = data.unsqueeze(0).cuda()
        data = data.to(device=device, dtype = torch.float)
        output = torch.sigmoid(model(data)).data.cpu()
        return output.numpy()
```

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
# Uncomment to launch training
model, loss_history = train_model(model,isbi, criterion, optimizer, n_epoch=60);
```
